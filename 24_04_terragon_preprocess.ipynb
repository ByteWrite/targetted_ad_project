```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import seaborn as sns # data visualization library  
import matplotlib.pyplot as plt

%matplotlib inline 
import warnings                        # To ignore any warnings warnings.filterwarnings("ignore")
```

```python
data = pd.read_csv('Terragon Data-2.csv')
data
```

```python
data.describe()
```

```python
# y includes our labels and x includes our features
#y = data.event_type                          # M or B 
list = ['msisdn','location_city','ad_id', 'ad_name','location_lga',
        'os_version','@timestamp']
data = data.drop(list,axis = 1 )
data
```

## Addressing missing values and Outlier

```python
data.isnull().sum()
```

#### Imputing missing values

```python
data['location_region'].fillna('unspecified',inplace=True)
data['location_state'].fillna('unspecified',inplace=True)
data['os_name'].fillna(value='unspecified', inplace=True)
data['customer_value'].fillna(value='unspecified', inplace=True)
data['gender'].fillna(value='unspecified', inplace=True)
data['device_type'].fillna(value='unspecified', inplace=True)
data['device_manufacturer'].fillna(value='unspecified', inplace=True)
```

```python
data['spend_total'].fillna(data['spend_total'].median(), inplace=True)
data['spend_vas'].fillna(data['spend_vas'].median(), inplace=True)
data['spend_voice'].fillna(data['spend_voice'].median(), inplace=True)
data['spend_data'].fillna(data['spend_data'].median(), inplace=True)
data['sms_cost'].fillna(value=0, inplace=True)
data['xtra_data_talk_rev'].fillna(data['xtra_data_talk_rev'].median(), inplace=True)
data['customer_class'].fillna(data['customer_class'].median(), inplace=True)

data['age'].replace(-70, 70,inplace=True)
data['age'].replace(-71, 71,inplace=True)
data['age'].replace(1051, 51,inplace=True)
data['age'].fillna(data['age'].median(), inplace=True)
```

```python
data['event_type'].fillna(data['event_type'].mode()[0], inplace=True)
```

```python
data['location_state'].value_counts()
```

### Outliers

```python
for j in data['age']:
    if j > 100:
        data['age'].replace(j,100,inplace=True)
    elif j < 12:
        data['age'].replace(j,11,inplace=True)
```

#### Categorical variables

```python
for i in data['os_name']:
    if i == 'android':
        i = i
    elif i == 'proprietary os':
        i= i
    else:
        data['os_name'].replace(i,'others',inplace=True)
```

```python
for j in data['device_manufacturer']:
    if j == 'tecno':
        j = j
    elif j == 'itel':
        j= j
    elif j == 'infinix':
        j= j
    elif j == 'samsung':
        j= j
    elif j == 'nokia':
        j= j
    elif j == 'apple':
        j= j
    else:
        data['device_manufacturer'].replace(j,'others',inplace=True)
```

```python
for k in data['event_type']:
    if k == 'sms':
        data['event_type'].replace(k,0,inplace=True)
    else:
        data['event_type'].replace(k,1,inplace=True)
```

```python
for l in data['location_state']:
    if l == 'lagos':
        l = l
    elif l == 'unspecified':
        l= l
    elif l == 'ogun':
        l= l
    elif l == 'rivers':
        l= l
    elif l == 'fct':
        l= l
    elif l == 'kano':
        l= l
    else:
        data['location_state'].replace(l,'others',inplace=True)
```

```python
data['location_state'].value_counts()
```

```python
for l in data['location_state']:
    if l == 'lagos' :
        data['location_state'].replace(l,37,inplace=True)
    elif l == 'unspecified':
        data['location_state'].replace(l,36,inplace=True)
    elif l == 'ogun':
        data['location_state'].replace(l,35,inplace=True)
    elif l == 'rivers' :
        data['location_state'].replace(l,34,inplace=True)
    elif l == 'fct':
        data['location_state'].replace(l,33,inplace=True)
    elif l == 'kano' :
        data['location_state'].replace(l,32,inplace=True)
    elif l == 'delta':
        data['location_state'].replace(l,31,inplace=True)
    elif l == 'oyo':
        data['location_state'].replace(l,30,inplace=True)
    elif l == 'kaduna':
        data['location_state'].replace(l,29,inplace=True)
    elif l == 'kogi' :
        data['location_state'].replace(l,28,inplace=True)
    elif l == 'abia':
        data['location_state'].replace(l,27,inplace=True)
    elif l == 'imo' :
        data['location_state'].replace(l,26,inplace=True)
    elif l == 'anambra':
        data['location_state'].replace(l,25,inplace=True)
    else:
        data['location_state'].replace(l,0,inplace=True)
        
        
#just tried this to see if it would give a better performance
```

#### Ordinal Variable 
There is a clear ordering in the Customer_value variable

```python
for l in data['customer_value']:
    if l == 'low' :
        data['customer_value'].replace(l,1,inplace=True)
    elif l == 'medium':
        data['customer_value'].replace(l,2,inplace=True)
    elif l == 'high' :
        data['customer_value'].replace(l,3,inplace=True)
    elif l == 'very high' :
        data['customer_value'].replace(l,4,inplace=True)
    elif l == 'top' :
        data['customer_value'].replace(l,5,inplace=True)
    else:
        data['customer_value'].replace(l,0,inplace=True)
```
